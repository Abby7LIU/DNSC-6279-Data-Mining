## Data Mining Assignment 3

```python
import pandas as pd
```

```python
# set measurement levels
col_types = {'chain': 'object',
             'offer': 'object',
             'market': 'object',
             'category': 'object',
             'company': 'object',
             'brand': 'object',
             'exact_item_bought': 'object'}
```

```python
path = 'assign_05_train.csv'
train = pd.read_csv(path,dtype=col_types)
train.shape
```

```python
path = 'assign_05_test.csv'
test = pd.read_csv(path,dtype=col_types)
test.shape
```

```python
path = 'assign_05_valid.csv'
valid = pd.read_csv(path,dtype=col_types)
valid.shape
```

### Loading into h2o

```python
import h2o
h2o.init()

SEED = 12345
```

```python
col_types = {'chain': 'enum',
             'offer': 'enum',
             'market': 'enum',
             'category': 'enum',
             'company': 'enum',
             'brand': 'enum',
             'exact_item_bought': 'enum'}

train_h2o = h2o.H2OFrame(train, column_types=col_types)
test_h2o = h2o.H2OFrame(test, column_types=col_types)
valid_h2o = h2o.H2OFrame(valid, column_types=col_types)
```

```python
train_h2o.describe()
```

### Target Encoding

```python
from h2o.targetencoder import TargetEncoder
y = 'repeater'
e_columns = ['market','chain']
te_= TargetEncoder(x=e_columns, y=y)
```

```python
# 'train' set
train_h2o[y]=train_h2o[y].asfactor()
_=te_.fit(train_h2o)
e_train=te_.transform(frame=train_h2o,holdout_type='loo',seed=SEED)
e_train.describe()
```

```python
# 'test' set
test_h2o[y]=test_h2o[y].asfactor()
e_test=te_.transform(frame=test_h2o,holdout_type='loo',seed=12345)
e_test.describe()
```

```python
# 'valid' set
valid_h2o[y]=valid_h2o[y].asfactor()
e_valid=te_.transform(frame=valid_h2o,holdout_type='loo',seed=12345)
e_valid.describe()
```

### GBM Tuning

```python
import numpy as np
import math
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.grid.grid_search import H2OGridSearch
```

```python
# check for the reponse and predictors
drops = ['id', 'chain', 'market', 'offerdate', 'quantity']
y = 'repeater'
X = [name for name in e_train.columns if name not in [y] + drops]
print(y)
print(X)
```

```python
# Set hyper parameters for GridSearch
hyper_params = {'ntrees': list(range(30,50,5)),
                'max_depth' : list(range(1,10,2)),
               'sample_rate': [0.8, 0.9],
               'col_sample_rate': [0.8, 0.9]}

Initial_GBM = H2OGradientBoostingEstimator(seed = SEED,
                                           score_tree_interval = 10, 
                                           stopping_rounds = 5,
                                           stopping_metric = "AUC",
                                           stopping_tolerance = 1e-4)

grid = H2OGridSearch(Initial_GBM, hyper_params,
                     grid_id = 'gbm',
                     search_criteria = {'strategy': "Cartesian"})
grid.train(x=X, 
           y=y,
           training_frame = e_train,
           validation_frame = e_valid)
```

```python
sorted_grid = grid.get_grid(sort_by='auc',decreasing=True)
print(sorted_grid)
```

```python
best_gbm = h2o.get_model(sorted_grid.sorted_metric_table()['model_ids'][0])
performance_best_gbm = best_gbm.model_performance(e_test)
print(performance_best_gbm.auc)
```

### MLP Modeling

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
hyper_parameters = {'hidden':[[170, 320], [80, 190], [320, 160, 80], [100], [50, 50, 50, 50]],
                    'l1':[s/1e4 for s in range(0, 1000, 100)],
                    'l2':[s/1e5 for s in range(0, 1000, 100)],
                    'input_dropout_ratio':[s/1e2 for s in range(0, 20, 2)]}

# define search strategy
search_criteria = {'strategy':'RandomDiscrete',           
                   'max_models':75,
                   'max_runtime_secs':900}

# initialize grid search
gsearch = H2OGridSearch(H2ODeepLearningEstimator,
                        hyper_params=hyper_parameters,
                        search_criteria=search_criteria)

# execute training w/ grid search
gsearch.train(x=X,
              y=y,
              training_frame=e_train,
              validation_frame=e_valid)
```

```python
sorted_gsearch = gsearch.get_grid(sort_by='auc',decreasing=True)
print(sorted_gsearch)
```

```python
best_mlp = h2o.get_model(sorted_gsearch.sorted_metric_table()['model_ids'][0])
performance_best_mlp = best_mlp.model_performance(e_test)
print(performance_best_mlp.auc)
```
